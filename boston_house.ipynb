## 윈저라이징

```python
from scipy.stats.mstats import winsorize

# 윈저라이징 함수 생성
def winsorize_col(winsordf, col):
    winsordf[col] = winsorize(winsordf[col], limits=[0.025, 0.025])
```

```python
# 윈저라이징할 컬럼으로만 구성된 데이터프레임
wins_boston = boston_feature.copy()
wins_boston.head()
```

## K-S Test

```python
import numpy as np
from scipy.stats import kstest

def ks_test(df_name):
    ks_columns = []
    ks_columns.append([col for col in df_name])
    for i in ks_columns:
        for j in i:
            ks_test = kstest(df_name[j], 'norm')
            print(j, "의 pvalue:", ks_test.pvalue)
```

```python
ks_test(wins_boston)
```

## Shapiro-Wilk Test

```python
# pvalue 함수

from scipy import stats
from scipy.stats import shapiro

# 등락률 컬럼만 추출해서 pvalue 구하는 함수
def shap_pct(df_name):
    pct_columns = []
    pct_columns.append([col for col in df_name])
    for i in pct_columns:
        for j in i:
            shapiro_test = shapiro(df_name[j])
            print(j, "의 pvalue:", shapiro_test.pvalue)
```

```python
shap_pct(wins_boston)
```

## VIF

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

def feature_engineering_XbyVIF(df_cols):
    vif = pd.DataFrame()
    vif['VIF_Factor'] = [variance_inflation_factor(df_cols.values, i)
                        for i in range(df_cols.shape[1])]
    vif['Feature'] = df_cols.columns
    return vif
```

```python
feature_vif = feature_engineering_XbyVIF(wins_boston)
feature_vif_df = pd.DataFrame(feature_vif)
feature_vif_df.sort_values('VIF_Factor', ascending=False).reset_index(drop=True)
```

## Bartlett 등분산 이분산 확인

```python
from scipy.stats import bartlett

homosced = []
heterosced = []

low_df = wins_boston_tg[wins_boston_tg['target'] == 0]   # low price house 
high_df = wins_boston_tg[wins_boston_tg['target'] == 1]  # high price house

bart_sced_df = pd.DataFrame(index=['등분산성'])

for i in wins_boston.columns:
    bart_sced_df[i] = ''
    
    ser_1 = low_df[i]
    ser_2 = high_df[i]

    bartresult = np.round(bartlett(ser_1, ser_2)[1], 3)
    
    if bartresult < 0.05:
        homosced.append(i)
        bart_sced_df.loc['등분산성', i] = '등분산'
        print(i,"pvalue", bartresult,"-> 등분산")
    else:
        heterosced.append(i)
        bart_sced_df.loc['등분산성', i] = '이분산'
        print(i,"pvalue", bartresult,"-> 이분산")
        
    print("**********************************************************")
```

```python
# 피처별 등분산성 데이터프레임 생성
bart_sced_df = bart_sced_df.transpose()
bart_sced_df
```

```python
bart_sced_df.value_counts()
```

## 등분산성을 가정한 Welch ttest

```python
import numpy as np
from scipy.stats import ttest_ind

low_homosced_df = wins_boston_tg.loc[:, homosced][wins_boston_tg['target'] == 0]    
high_homosced_df = wins_boston_tg.loc[:, homosced][wins_boston_tg['target'] == 1]  

significant_homosced = []       # 유의수준 피처

# 등분산성 Welch's T-test
def ttest_homosced(df_name1, df_name2):
    for i in low_homosced_df.columns:
        welch_result = ttest_ind(df_name1[i], df_name2[i], equal_var=True)

        if welch_result.pvalue < 0.05:
            significant_homosced.append(i)

        print("등분산을 가정한", i, "의 pvalue:", np.round(welch_result.pvalue, 3))
    
        print("**********************************************************")
```

```python
ttest_homosced(low_homosced_df, high_homosced_df)
```

```python
# 유의수준 피처
print('유의수준 피처 개수: ', len(significant_homosced), '개')
print(significant_homosced)
```

## 이분산을 가정한 Welch ttest

```python
import numpy as np
from scipy.stats import ttest_ind

low_heterosced_df = wins_boston_tg.loc[:, heterosced][wins_boston_tg['target'] == 0]    
high_heterosced_df = wins_boston_tg.loc[:, heterosced][wins_boston_tg['target'] == 1]  

significant_heterosced = []       # 유의수준 피처

# 등분산성 Welch's T-test
def ttest_heterosced(df_name1, df_name2):
    for i in low_heterosced_df.columns:
        welch_result = ttest_ind(df_name1[i], df_name2[i], equal_var=False)

        if welch_result.pvalue < 0.05:
            significant_heterosced.append(i)

        print("이분산을 가정한", i, "의 pvalue:", np.round(welch_result.pvalue, 3))
    
        print("**********************************************************")
```

```python
ttest_heterosced(low_heterosced_df, high_heterosced_df)
```

```python
# 유의수준 피처
print('유의수준 피처 개수: ', len(significant_heterosced), '개')
print(significant_heterosced)
```

```python
print('등분산 유의수준 피처: ', significant_homosced)
print('-------------------------------------------------------------------')
print('이분산 유의수준 피처: ', significant_heterosced)
```

## Levene

```python
from scipy.stats import levene

homosced = []
heterosced = []

lev_df = wins_boston_tg.iloc[:, :-1]

nc_df = wins_boston_tg[wins_boston_tg['target'] ==0]  
bc_df = wins_boston_tg[wins_boston_tg['target'] ==1]  

lev_sced_df = pd.DataFrame(index=['등분산성'])

for i in lev_df.columns:
    lev_sced_df[i] = ''
    
    ser_1 = nc_df[i]
    ser_2 = bc_df[i]

    lresult = np.round(levene(ser_1, ser_2)[1], 3)
    
    if lresult < 0.05:
        homosced.append(i)
        lev_sced_df.loc['등분산성', i] = '등분산'
        print(i,"pvalue", lresult,"-> 등분산")
    else:
        heterosced.append(i)
        lev_sced_df.loc['등분산성', i] = '이분산'
        print(i,"pvalue", lresult,"-> 이분산")

    print("**********************************************************")
```

```python
# 피처별 등분산성 데이터프레임 생성
lev_sced_df = lev_sced_df.transpose()
lev_sced_df
```
