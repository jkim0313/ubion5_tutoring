csv 파일을 불러오세요.

```python
df1 = pd.read_csv('./pandas_practice/csv/2014.csv', encoding='CP949')
df2 = pd.read_csv('./pandas_practice/csv/2015.csv', encoding='CP949')
df3 = pd.read_csv('./pandas_practice/csv/2016.csv', encoding='CP949')
```

concat을 사용해서 하나의 데이터프레임으로 병합하세요.
인덱스를 재설정하세요.

```python
dfs_to_concat = [df1, df2, df3]

dfs = pd.concat(dfs_to_concat)
dfs.reset_index(inplace=True, drop=True)
dfs
```

```python
dfs.head()
```

"구분" 컬럼에서 "년"을 "-"로 대체하세요.
"구분" 컬럼에서 "월"을 공백으로로 대체하세요.

```python
dfs['구분'] = dfs['구분'].str.replace('년', '-')
dfs['구분'] = dfs['구분'].str.replace('월', '')
dfs['구분'].head(5)
```

"구분" 컬럼을 datetime 형식으로 바꾸세요.

```python
dfs["구분"] = pd.to_datetime(dfs["구분"])     # 구분 컬럼을 datetime 형식으로 타입 변경
dfs["구분"] = dfs["구분"].dt.strftime("%Y%m")                    # 년월 포맷 설정

dfs["구분"].head()
```

```python
dfs.info()
```

"년도" 라는 이름의 컬럼을 추가하고 "구분" 컬럼에서 년도만 따로 추출해서 입력하세요.

```python
dfs["년도"] = dfs["구분"].dt.strftime("%Y")   # 년도 컬럼에 구분 컬럼의 년도만 따로 추출해서 입력

dfs.head()
```

년도별 평균 사망자 수와 총 사망자 수를 출력하세요.

```python
dfs.groupby("년도")["사망(명)"].mean()
```

```python
dfs.groupby("년도").agg(["mean", "sum"])["사망(명)"]   # 년도별 평균 사망자 수, 총 사망자 수 출력
```

년도별 사고대비 사망율을 출력하세요.

```python
accident = dfs.groupby("년도").sum()["사고(건)"]   # 년도별 총 사고건 수
death = dfs.groupby("년도").sum()["사망(명)"]      # 년도별 총 사망자 수

death_rate = (accident/death)                           # 사고대비 사망율
death_rate
```
